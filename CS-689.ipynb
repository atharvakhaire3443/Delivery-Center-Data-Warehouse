```python
# Function for Date Transformation

from datetime import datetime

def extract_components(date_time_str):
    if pd.isna(date_time_str):  # handle NaN values
        return pd.Series({
            'month': '',
            'day': '',
            'year': '',
            'hour': '',
            'minute': ''
        })
    else:
        try:
            date_time_obj = datetime.strptime(date_time_str, '%m/%d/%Y %H:%M')
            return pd.Series({
                'month': date_time_obj.month,
                'day': date_time_obj.day,
                'year': date_time_obj.year,
                'hour': date_time_obj.hour,
                'minute': date_time_obj.minute
            })
        except ValueError:
            # handle invalid date-time strings
            return pd.Series({
                'month': '',
                'day': '',
                'year': '',
                'hour': '',
                'minute': ''
            })

# Apply function to date_time column and create new columns for each component
orders_staging['order_moment_accepted'] = orders_staging['order_moment_accepted'].astype(str)
orders_staging[['order_moment_accepted_month', 'order_moment_accepted_day', 'order_moment_accepted_year', 'order_moment_accepted_hour', 'order_moment_accepted_minute']] = orders_staging['order_moment_accepted'].apply(extract_components)

# Print the updated DataFrame
print(orders_staging)
```
